```python
#creating function and appending customer and order info to it.
def agg_customer(x):
    d = []
    d.append(x['Order ID'].count())
    d.append(x['Sales'].sum())
    d.append(x['Profit %'].mean())
    d.append(pd.to_datetime(x['Order Date']).min())
    d.append(pd.to_datetime(x['Order Date']).max())
    d.append(x['Product Name'].unique())
    d.append(x['City'].unique())
    return pd.Series(d, index=['#Purchases','Total_Sales','Average Profit % gained','First_Purchase_Date','Latest_Purchase_Date','Products Purchased','Location_Count'])
```

```python
#grouping based on Customer ID and applying the function we created above
df_agg = df.groupby('Customer ID').apply(agg_customer)
df_agg
```

## Time Series Forcasting

```python
furniture= df.loc[df['Category']=='Furniture']
```

```python
furniture['Order Date'].min() #min date in dataset
```

```python
furniture['Order Date'].max() #max date in dataset
```

```python
cols = ['Order ID','Ship Date','Customer ID','Ship Mode','Customer Name','Segment',
       'City','State','Postal Code','Region','Product ID','Category','Sub-Category'
       ,'Product Name','Quantity','Discount','Profit']
```

```python
furniture.drop(cols,axis=1,inplace = True)
```

```python
furniture.head()
```

```python
furniture.drop(['Cost','Profit %','Shipment Duration','order year'],axis=1,inplace = True)
```

```python
furniture.head()
```

```python
furniture.sort_values('Order Date')
furniture.isnull().sum()
```

```python
furniture = furniture.groupby('Order Date')['Sales'].sum().reset_index()
```

```python
furniture = furniture.set_index('Order Date') #indexing with time series data
```

```python
furniture.index[0:5]
```

```python
y = furniture['Sales'].resample('MS').mean()
```

```python
y['2017':][0:5]
```

```python
y.plot(figsize = (15,6))
plt.show()
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 18,8

decomposition = sm.tsa.seasonal_decompose(y,model = 'additive')
fig = decomposition.plot()
plt.show()
```

```python
#time series forcasting model using ARIMA
p = d = q = range(0,2)
pdq = list(itertools.product(p,d,q))
```

```python
seasonal_pdq = [(x[0],x[1],x[2],12) for x in list(itertools.product(p,d,q))] #list comprehension
```

```python
print('Examples of parameter combination for seasonal ARIMA')
print('SARIMAX: {} x {}'.format(pdq[1],seasonal_pdq[1]))
print('SARIMAX: {} x {}'.format(pdq[2],seasonal_pdq[2]))
print('SARIMAX: {} x {}'.format(pdq[3],seasonal_pdq[3]))
```

```python
pdq
```

```python
seasonal_pdq
```

```python
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(y,order = param,seasonal_order = param_seasonal,
                                           enforce_stationarity=False,enforce_invertibility=False)
            results = mod.fit()
            print('ARIMA {} x {} 12 - AIC {}'.format(param,param_seasonal,results.aic))
        except:
            continue
```

```python
#ARIMA (0,1,1)x(0,1,1,12) hyperparameter combination giving the best results with min AIC
#fitting the ARIMA model
#ARIMA (0,1,1)
mod = sm.tsa.statespace.SARIMAX(y,order = (0,1,1),seasonal_order=(0,1,1,12),
                               enforce_stationarity=False,enforce_invertibility=False)
results=mod.fit()
print(results.summary().tables[1])
```

```python
results
```

```python
#running model diagnostics using graph
results.plot_diagnostics(figsize = (15,6))
plt.show()
```

```python
#build validated forcasts
pred = results.get_prediction(start=pd.to_datetime('2017-01-01'),dynamic = False)
```

```python
pred
```

```python
pred_ci = pred.conf_int()#getting confidence interval for furniture sales
```

```python
pred_ci
```

```python
ax = y['2014':].plot(label = 'observed')
pred.predicted_mean.plot(ax = ax , label = 'one step ahead forecast', alpha = 0.7,figsize = (14,7))
ax.fill_between(pred_ci.index,pred_ci.iloc[:,0],pred_ci.iloc[:,1],color = 'k' , alpha= 0.2)
ax.set_xlabel('Date')
ax.set_ylabel('Furniture Sales')
plt.legend()
plt.show()
```
